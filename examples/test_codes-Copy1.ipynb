```python
from turboeda import EDAReport
```

```python
input_path = "ncr_ride_bookings.csv" # or xlsx
```

```python
report = EDAReport(
    input_path=input_path,
    theme="dark",                # or "light"
    # sheet="Sheet1",            # optional; if missing, we use the first sheet
    sample_rows=None,            # all rows or sample, default=200000
    # max_numeric_plots=12,      # default=12
    # max_categorical_plots=12   # default=12
    max_corr_cols=100000         # default=40
)
```

```python
report = EDAReport(
    input_path=input_path,
    theme="dark",
    # sample_rows=None,
    max_corr_cols=100_000,
    auto_save_and_open=True,     # <-- NEW: save & open right after run()
    # out_path="bank_report.html",  # optional custom name; otherwise <input>_report.html
)

res = report.run()               # computes EDA, saves HTML and opens it in a new browser tab
```
